## Q 1a) Find the frequencies of POS tags of words.

## Q 1f) Find the directed POS tag tuples and For each such 2-tuple, list the frequencies separately for each relation R as well as total.




```python

## finding the frequencies of relation for each 2 tuple.
from collections import OrderedDict
from operator import getitem

dict_nest={}
for i in range(len(sentences)):
    sentence=sentences[i]
    for token in sentence:
        id=token['head']
        for innertoken in sentence:
            if(innertoken['id']==id):
                t1=(token['upos'],innertoken['upos'])
                if(t1  in dict_nest):
                    if(token['deprel'] in dict_nest[t1] ):
                        dict_nest[t1][token['deprel']]=dict_nest[t1][token['deprel']]+1
                    else:
                        dict_nest[t1][token['deprel']]=1
                else:
                    dict_nest[t1]={}

print("the frequncies for each relation:")

print(dict_nest)
    


### finding the frequencies of relation in total.



sum=0
dict_nest2={}
for key,value in dict_nest.items():
    n=key
    for key,value in value.items():
         sum=sum+value   
    dict_nest2[n]=sum

print("The total frequencies:")
dict_nest2= {k: v for k, v in sorted(dict_nest2.items(), key=lambda item: item[1],reverse=True)}

print(dict_nest2)  






```
